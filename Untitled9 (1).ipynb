```python
#problem statement
#An e-commerce company wants to understand its sales performance, customer behaviour, product trends, and payment preferences across different regions.
#The goal of this project is to analyze transactional data to identify revenue drivers, improve category-level performance, optimize inventory, and improve customer retention.
#Using real-time e-commerce data, I performed data cleaning, exploratory data analysis (EDA), and extracted actionable insights that can support marketing, product, and business teams in decision-making.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (10,5)

from ipywidgets import interact, Dropdown
```

```python

df = pd.read_csv("D:\ecommerce_dataset_updated.csv")
df.head()
```

```python
df.info()
```

```python
# Rename columns properly
df.columns = ['User_ID', 'Product_ID', 'Category', 'Price', 'Discount', 'Final_Price',
              'Payment_Method', 'Purchase_Date']
```

```python
# Convert date column
df['Purchase_Date'] = pd.to_datetime(df['Purchase_Date'], errors='coerce')
```

```python
# Handle missing values
df.fillna({
    'Category': 'Unknown',
    'Payment_Method': 'Unknown'
}, inplace=True)
```

```python
# If any numerical NA
df['Price'] = df['Price'].fillna(df['Price'].median())
df['Final_Price'] = df['Final_Price'].fillna(df['Final_Price'].median())
df['Discount'] = df['Discount'].fillna(0)
```

```python
# Create new helpful columns
df['Month'] = df['Purchase_Date'].dt.month_name()
df['Year'] = df['Purchase_Date'].dt.year
df['Day'] = df['Purchase_Date'].dt.day_name()
```

```python
df.head()
```

```python
#EDA
#Basic Stats
df.describe(include='all')
```

```python
#Total Revenue
total_revenue = df['Final_Price'].sum()
total_revenue
```

```python
#Category Distribution
df['Category'].value_counts().plot(kind='bar')
plt.title("Product Category Count")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
#Revenue by Category
category_revenue = df.groupby('Category')['Final_Price'].sum().sort_values(ascending=False)

category_revenue.plot(kind='bar')
plt.title("Revenue by Category")
plt.ylabel("Revenue")
plt.show()
```

```python
#Payment Method Distribution
df['Payment_Method'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Payment Method Split")
plt.ylabel("")
plt.show()

```

```python
#Monthly Sales Trend
monthly_sales = df.groupby('Month')['Final_Price'].sum().reindex([
    'January','February','March','April','May','June',
    'July','August','September','October','November','December'
])

monthly_sales.plot(marker='o')
plt.title("Monthly Sales Trend")
plt.ylabel("Sales")
plt.show()

```

```python
#Discount vs Final Price
sns.scatterplot(data=df, x='Discount', y='Final_Price')
plt.title("Discount vs Final Price")
plt.show()
```

```python
print(" Business Insights\n")
print(f"Total Revenue: {total_revenue:,.2f}\n")

print("Top 5 Selling Categories:")
print(df['Category'].value_counts().head(), "\n")

print("Top 5 Categories by Revenue:")
print(category_revenue.head(), "\n")

print("Most Used Payment Method:")
print(df['Payment_Method'].value_counts().idxmax(), "\n")

print("Best Month for Sales:")
print(monthly_sales.idxmax(), "\n")

print("Highest Spending User:")
print(df.groupby('User_ID')['Final_Price'].sum().idxmax(), "\n")
```
